```python
colunas = ['SECRETARIA RESPONSÁVEL', 'GRUPO DE SERVIÇO', 'NOME DO SERVIÇO',
           'CLIENTE DO SERVIÇO (Diga para quem esse serviços se destina)',
           'O QUE É O SERVIÇO (Descreva o objetivo do serviço)',
           'DOCUMENTAÇÃO NECESSÁRIA (faça uma listagem documentação necessaria para solicitar o serviço)',
           'LEGISLAÇÃO APLICADA']

for coluna in colunas:
    listagem[coluna] = listagem[coluna].str.replace('[()§] ','',regex=True)
    listagem[coluna] = listagem[coluna].str.replace('\n','',regex=True)

#listagem['DOCUMENTAÇÃO NECESSÁRIA (faça uma listagem documentação necessaria para solicitar o serviço)'] = listagem['DOCUMENTAÇÃO NECESSÁRIA (faça uma listagem documentação necessaria para solicitar o serviço)'].str.replace('[()§]\n','',regex=True)
listagem['DOCUMENTAÇÃO NECESSÁRIA (faça uma listagem documentação necessaria para solicitar o serviço)']
```
